```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
df = pd.read_csv('cars_Cleaned.csv')
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.iloc[:,:10].corr(), annot = True)
```

```python
df.iloc[:,:10].hist(figsize=(10,8))
plt.tight_layout()
```

```python
df_copy = df.copy()

# Mapping Fuel
fuel_map = {1: 'Gasolina', 0: 'Diesel'}
df_copy['Fuel'] = df_copy['Fuel'].map(fuel_map)

fig = px.scatter(df_copy, x = 'Mileage', y = 'price', color='Fuel',hover_data='power')

fuel_types = df_copy['Fuel'].unique()
buttons = []

# Botón para mostrar todos los datos
df_copy['Fuel'] = df_copy['Fuel'].astype(str)
buttons.append(dict(
    args=[{'visible': [True] * len(fig.data)}],
    label='Todos',
    method='update'
))

# Botones para cada tipo de combustible
for fuel in fuel_types:
    buttons.append(dict(
        args=[{'visible': [fuel == trace.name for trace in fig.data]}],
        label=fuel,
        method='update'
    ))

# Añadir menú de botones al gráfico
fig.update_layout(
    updatemenus=[dict(
        type="buttons",
        direction="right",
        buttons=buttons,
        pad={"r": 10, "t": 10},
        showactive=True,
        x=1.02,
        xanchor="left",
        y=0.8,
        yanchor="top"
    )]
)

# Ajustar el diseño del gráfico
fig.update_layout(
    title='Precio vs. Kilometraje por Tipo de Combustible',
    xaxis_title='Kilometraje',
    yaxis_title='Precio',
    legend_title='Tipo de Combustible',
    height = 800
)

fig.show()
```

```python
# Seleccionar las columnas que comienzan con 'Comfort_'
comfort_columns = [col for col in df.columns if col.startswith('Comfort_')]

# Calcular el precio promedio general
average_price = df['price'].mean()

# Calcular el precio promedio para cada característica de confort
comfort_prices = []
for col in comfort_columns:

    #Selecciona solo las filas donde la característica está presente y calcula el precio promedio de esos vehículos
    avg_price_with_feature = df[df[col] == 1]['price'].mean()

    #Calcula cuánto más caros son en promedio los vehículos con esta característica comparados con el precio promedio general
    price_difference = avg_price_with_feature - average_price
    comfort_prices.append((col.replace('Comfort_', ''), price_difference))

# Ordenar las características por diferencia de precio y seleccionar las top 10
# El lambda indica que se debe ordenar basándose en el segundo elemento de cada tupla (la diferencia de precio)
top_10_comfort = sorted(comfort_prices, key=lambda x: x[1], reverse=True)[:10]

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x=[item[0] for item in top_10_comfort], y=[item[1] for item in top_10_comfort])
plt.title('Top 10 Características de Confort que Más Influyen en el Precio', fontsize=16)
plt.xlabel('Características de Confort', fontsize=12)
plt.ylabel('Diferencia de Precio (€)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
gearbox_map = {0: 'Automático', 1: 'Manual'}
df['Gearbox'] = df['Gearbox'].map(gearbox_map)
gearbox_counts = df['Gearbox'].value_counts()

fig = px.pie(
    values=gearbox_counts.values,
    names=gearbox_counts.index,
    title='Tipos de Transmisión'
)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
df = pd.read_csv('cars_Cleaned.csv')
# Mapear los valores de combustible a etiquetas más descriptivas
fuel_map = {0: 'Gasolina', 1: 'Diesel'}
df['Fuel'] = df['Fuel'].map(fuel_map)

# Crear el gráfico de caja y bigotes
fig = px.box(df, x='Fuel', y='price', 
             title='Distribución de Precios por Tipo de Combustible',
             labels={'Fuel': 'Tipo de Combustible', 'price': 'Precio (€)'},
             color='Fuel',
             points="all",
             category_orders={'Fuel': ['Gasolina', 'Diesel']})

# Personalizar el diseño
fig.update_layout(
    xaxis_title='Tipo de Combustible',
    yaxis_title='Precio (€)',
    showlegend=False,
    height = 800
)

# Mostrar el gráfico
fig.show()
```

```python
df = pd.read_csv('cars_Cleaned.csv')

# Mapear los valores de combustible a etiquetas más descriptivas
fuel_map = {0: 'Gasolina', 1: 'Diesel'}
df['Fuel'] = df['Fuel'].map(fuel_map)

# Crear el gráfico de dispersión 3D
fig = px.scatter_3d(df, x='Mileage', y='power', z='price',
                    color='Fuel',
                    labels={'Mileage': 'Kilometraje', 
                            'power': 'Potencia (HP)', 
                            'price': 'Precio (€)',
                            'Fuel': 'Combustible'},
                    title='Precio vs. Kilometraje vs. Potencia')

# Ajustar el diseño
fig.update_layout(scene = dict(
                    xaxis_title='Kilometraje',
                    yaxis_title='Potencia (HP)',
                    zaxis_title='Precio (€)'),
                  width=900, height=700,
                  margin=dict(r=20, b=10, l=10, t=40))

# Mostrar el gráfico
fig.show()
```
